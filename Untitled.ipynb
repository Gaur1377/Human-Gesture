```python
import cv2
import numpy as np
import  os
from  matplotlib import pyplot as plt
import time
import mediapipe as mp
```

```python
mp_holistic = mp.solutions.holistic
mp_Drawing = mp.solutions.drawing_utils
```

```python
def mediapipe_detection(imge,model):
    imge = cv2.cvtColor(imge,cv2.COLOR_BGR2RGB)
    imge.flags.writeable = False
    results = model.process(imge)
    imge.flags.writeable = True
    imge = cv2.cvtColor(imge,cv2.COLOR_BGR2RGB)
    return imge,results
```

```python
def draw_detection(imge, results):
    # mp_Drawing.draw_landmarks(imge,results.face_landmarks,mp_holistic.FACEMESH_TESSLATION)
    mp_Drawing.draw_landmarks(imge,results.pose_landmarks,mp_holistic.POSE_CONNECTIONS)
    mp_Drawing.draw_landmarks(imge,results.left_hand_landmarks,mp_holistic.HAND_CONNECTIONS)
    mp_Drawing.draw_landmarks(imge,results.right_hand_landmarks,mp_holistic.HAND_CONNECTIONS)
```

```python
def draw_style_landmarks(imge,results):
    mp_Drawing.draw_landmarks(imge,results.pose_landmarks,mp_holistic.POSE_CONNECTIONS,
                             mp_Drawing.DrawingSpec(color=(80,22,10),thickness=2,circle_radius=4),
                              mp_Drawing.DrawingSpec(color=(80,44,121),thickness=2,circle_radius=1))
    mp_Drawing.draw_landmarks(imge,results.left_hand_landmarks,mp_holistic.HAND_CONNECTIONS,
                             mp_Drawing.DrawingSpec(color=(80,22,10),thickness=2,circle_radius=4),
                              mp_Drawing.DrawingSpec(color=(80,44,121),thickness=2,circle_radius=2))
    mp_Drawing.draw_landmarks(imge,results.right_hand_landmarks,mp_holistic.HAND_CONNECTIONS,
                             mp_Drawing.DrawingSpec(color=(245,117,66),thickness=2,circle_radius=4),
                              mp_Drawing.DrawingSpec(color=(245,66,230),thickness=2,circle_radius=2))
```

```python
cap = cv2.VideoCapture(0)
with mp_holistic.Holistic(min_detection_confidence=0.5,min_tracking_confidence=0.5) as holistic:
    while cap.isOpened():
        ret, frame = cap.read()


        imge , results = mediapipe_detection(frame,holistic)
        print(results)
        draw_style_landmarks(imge,results)

        cv2.imshow("images",imge)
        if cv2.waitKey(1) & 0xFF ==ord('q'):
            break
    cap.release()
    cv2.destroyWindow()
```

```python
# import numpy as np

def extract_keypoint(results):
    # Pose: 33 landmarks, each with x, y, z, visibility
    pose = np.array([[res.x, res.y, res.z, res.visibility] 
                     for res in results.pose_landmarks.landmark]).flatten() if results.pose_landmarks else np.zeros(33 * 4)
    
    # Left hand: 21 landmarks, each with x, y, z
    lh = np.array([[res.x, res.y, res.z] 
                   for res in results.left_hand_landmarks.landmark]).flatten() if results.left_hand_landmarks else np.zeros(21 * 3)
    
    # Right hand: 21 landmarks, each with x, y, z
    rh = np.array([[res.x, res.y, res.z] 
                   for res in results.right_hand_landmarks.landmark]).flatten() if results.right_hand_landmarks else np.zeros(21 * 3)
    
    return np.concatenate([pose, lh, rh])
```

```python
DATA_PATH = os.path.join('mp_data')
actions = np.array(['hello','thank','Allthebest'])
no_sequence = 20
sequence_length = 20
```

```python
for action in actions:
    for sequence in range(no_sequence):
        try:
            os.makedirs(os.path.join(DATA_PATH,action,str(sequence)))
        except:
            pass
```

```python
cap = cv2.VideoCapture(0)
with mp_holistic.Holistic(min_detection_confidence=0.5,min_tracking_confidence=0.5) as holistic:
    # while cap.isOpened():
    for action in actions:
        for sequence in range(no_sequence):
            for frame_num in range(sequence_length):
                ret, frame = cap.read()

            
                imge , results = mediapipe_detection(frame,holistic)
                print(results)
                draw_style_landmarks(imge,results)
                if frame_num == 0:
                    cv2.putText(imge,'STARTING COLLECTING',(120,200),
                               cv2.FONT_HERSHEY_SIMPLEX,1,(0,255,0),4, cv2.LINE_AA)
                    cv2.putText(imge,'Collecting  frame for {} video Number {}'.format(action, sequence),(15,12),
                               cv2.FONT_HERSHEY_SIMPLEX,0.5,(0,0,255),1, cv2.LINE_AA)
                    cv2.imshow("images",imge)
                    cv2.waitKey(2000)
                else:
                    cv2.putText(imge,'Collecting  frame for {} video Number {}'.format(action, sequence),(15,12),
                               cv2.FONT_HERSHEY_SIMPLEX,0.5,(0,0,255),1, cv2.LINE_AA)
                    cv2.imshow("images",imge)

                keypoints = extract_keypoint(results)
                npy_path = os.path.join(DATA_PATH,action,str(sequence),str(frame_num))
                np.save(npy_path, keypoints)
                

                # cv2.imshow("images",imge)
                if cv2.waitKey(10) & 0xFF ==ord('q'):
                    break
   
    cap.release()
    cv2.destroyAllWindows() 
```

```python
from sklearn.model_selection import train_test_split
from tensorflow.keras.utils import to_categorical
```

```python
label_map = {label:num for num,label in enumerate(actions)}
```

```python
label_map
```

```python
sequences,labels = [], []
for action in actions:
    for sequence in range(no_sequence):
        window = []
        for frame_num in range(sequence_length):
           file_path = os.path.join(DATA_PATH, action, str(sequence), f'{frame_num}.npy')
           if os.path.exists(file_path):
             res = np.load(file_path)
             if res.shape == (258,):  # Ensure correct shape
                    window.append(res)
             else:
                    print(f"❌ Wrong shape in {file_path}: {res.shape}")
        if len(window) == sequence_length:
            sequences.append(window)
            labels.append(label_map[action])
```

```python
np.array(sequences).shape
```

```python
X = np.array(sequences)  # shape: (num_sequences, sequence_length, 258)
y = to_categorical(labels).astype(int)  # one-hot encoded labels
```

```python
X.shape
```

```python
np.array(labels).shape
```

```python
X_train,X_test,y_train,y_test = train_test_split(X,y,test_size = 0.05)
```

```python
X_test.shape
```

```python
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import LSTM,Dense
from tensorflow.keras.callbacks import TensorBoard
```

```python
log_dir = os.path.join('Logs')
tb_callback = TensorBoard(log_dir=log_dir)
```

```python
model = Sequential()
model.add(LSTM(64, return_sequences=True, activation='relu', input_shape=(30, 258)))
model.add(LSTM(128, return_sequences=True, activation='relu'))
model.add(LSTM(64, return_sequences=False, activation='relu'))
model.add(Dense(64, activation='relu'))
model.add(Dense(32, activation='relu'))
model.add(Dense(actions.shape[0], activation='softmax')) 
```

```python
model.compile(optimizer='Adam',loss = 'categorical_crossentropy', metrics = ['categorical_accuracy'])
```

```python
model.fit(X_train,y_train,epochs = 2000,callbacks = [tb_callback])
```
